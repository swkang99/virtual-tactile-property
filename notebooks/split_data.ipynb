```python
from pathlib import Path
data_original_path = Path('../data/original')
```

```python
train_cnt = 70
val_cnt = 20
test_cnt = 10
```

```python
import random
idx_list = list(range(1, 101))
random.shuffle(idx_list)
print(idx_list)
```

```python
train_list = idx_list[:train_cnt]
val_list = idx_list[train_cnt:train_cnt+val_cnt]
test_list = idx_list[val_cnt:val_cnt+test_cnt]
print(train_list)
print(val_list)
print(test_list)
```

```python
texture_path = data_original_path / Path('texture_image')
texture_list = [x.name for x in texture_path.iterdir() ]
print(texture_list)
```

```python
import shutil
data_split_path = Path('../data/split')
split_train_path = Path('train/')
split_val_path = Path('val/')
split_test_path = Path('test/')

for texture in texture_list:
    src = texture_path / texture
    idx = int(src.stem)
    if idx in test_list:
        dst = data_split_path / split_test_path / Path('texture_image') / src.name
    elif idx in val_list:
        dst = data_split_path / split_val_path / Path('texture_image') / src.name
    else: # train
        dst = data_split_path / split_train_path / Path('texture_image') / src.name

    dst.parent.mkdir(parents=True, exist_ok=True)
    shutil.copy(src, dst)
```

```python
import pandas as pd
csv = data_original_path / Path('ParticipantData.csv')
df = pd.read_csv(csv, header=None)
df = df + 50 # Restore the value range to its original state  
df.head()
```

```python
roughness_list = df.iloc[:, 0].tolist()
```

```python
train_records = []
val_records = []
test_records = []

for r_idx in range(1, len(roughness_list) + 1):
    row = {
        'file_name': r_idx,
        'roughness': roughness_list[r_idx - 1]
    }
    if r_idx in test_list:
        test_records.append(row)
    elif r_idx in val_list:
        val_records.append(row)
    else:
        train_records.append(row)

train_df = pd.DataFrame(train_records)
val_df = pd.DataFrame(val_records)
test_df = pd.DataFrame(test_records)

train_df.to_csv(data_split_path / split_train_path / 'train.csv', index=False)
val_df.to_csv(data_split_path / split_val_path / 'val.csv', index=False)
test_df.to_csv(data_split_path / split_test_path / 'test.csv', index=False)
```
